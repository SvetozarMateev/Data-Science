```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
items =[8,3,1,8,6,5,10,1,8,1]
def Counts (items):
    countsOfItems={}
    for item in items:
        if not (item in countsOfItems):
            countsOfItems[item]=1
        else:
            countsOfItems[item]+=1
    for (item,count) in countsOfItems.items():
        print(item, count/len(items))
    return countsOfItems
print(Counts(items))

plt.hist(items)
plt.xlim(1,10)
plt.show()

plt.hist(items, normed=1)
plt.xlim(1,10)
plt.show()

plt.hist(items, cumulative= True)
plt.xlim(1,10)
plt.show()
```

```python
data_snowfall= pd.read_table('snowfall.csv', header= None)
data_snowfall.head()
plt.figure()
plt.hist(data_snowfall, bins= 15)
plt.axvline(data_snowfall[0].mean(), color='red', linewidth=2, label = 'Mean')
plt.axvline(data_snowfall[0].median(),color = 'green', linewidth=2, label = 'Median')
plt.legend()
plt.show()
```
