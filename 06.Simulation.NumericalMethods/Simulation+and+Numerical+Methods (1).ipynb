```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import math
import scipy.interpolate as interp
import scipy.optimize
```

```python
def f(x):
    return np.sin(x*x)
points_in_x=[]
points_in_y=[]
def integrate(x_max,x_min,points,y_max,y_min):
    points_in=0
    points_all= int(points)
    for i in range (points_all+1):
        for a in range(points_all+1):
            x=np.random.uniform(x_min,x_max)
            y = np.random.uniform(y_min,y_max)
            if f(x)>0 and y<=f(x) and y>0:
                points_in_x.append(x)
                points_in_y.append(y)
                points_in+=1
            if f(x)<0 and y>=f(x) and y<0:
                points_in_x.append(x)
                points_in_y.append(y)
                points_in-=1
                
    rectangle_area=math.fabs(y_max-y_min)*math.fabs(x_max-x_min)
    return(points_in/points_all**2*rectangle_area) 

print(integrate(math.pi,0,1000,1,-1))
plt.scatter(points_in_x,points_in_y)
plt.show()
```

```python
def f(x):
    return x**3-2*x-5
def df(x):
    return 3*x**2-2
root = scipy.optimize.newton(f, 1, df)
value = f(root) # Should be close to 0
print(root, value)
```

```python
def secondf(x):
    return np.tan(x)-4*x #tan(x)=4x
def seconddf(x):
    return 1/np.cos(x)**2-4
root = scipy.optimize.newton(secondf, 1.5, seconddf)
value = secondf(root) # Should be close to 0
print(root, value)
```

```python
def calculate_sqrt(x,guess=1,precision=1e-12):
    while not(abs(guess*guess-x)<precision):
        guess=(guess+(x/guess))/2.0
    return guess
print(calculate_sqrt(36))
print(calculate_sqrt(16))
print(calculate_sqrt(4))
print(calculate_sqrt(9))

    
```
